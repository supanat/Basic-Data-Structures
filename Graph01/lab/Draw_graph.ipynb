```python
graph ={'a':{'b','c','f'},
       'b':{'a','d','f'},
       'c':{'a','d','e'},
       'd':{'b','c','f'},
       'e':{'c','f'},
       'f':{'a','b','d','e'}}
```

```python
G = nx.Graph()

for n in graph.keys():
    G.add_node(n)

for i in graph:
    nodes = graph[i]
    from_node = i;
    #print(from_node,':')
    
    for successor in nodes:
        G.add_edge(from_node, successor)
        #print(successor)
        
nx.draw(G,with_labels=True);   
```

```python
def draw_graph(graph):
    options = {
        'node_color': 'yellow',
        'node_size': 600,
        'width': 1,
    }
    
    for n in graph.keys():
        G.add_node(n)

    for i in graph:
        nodes = graph[i]
        from_node = i;
        #print(from_node,':')
    
        for successor in nodes:
            G.add_edge(from_node, successor)
            #print(successor)    
    nx.draw(G,with_labels=True, **options);  
```

```python
draw_graph(graph)
```
